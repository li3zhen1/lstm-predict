```python
# convert df2["Date"] to datetime
df2["Date"] = pd.to_datetime(df2["Date"])
df2.dtypes
```

```python
# fill the predicted close price

for i in range(len(df)):
    df2.loc[df2["Date"] == df["Date"][i], "Predicted_Close"] = df["Predicted_Close"][i]

df2
```

```python
# plot "Close" and "Predicted_Close" in one graph
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10), dpi=100)
plt.plot(df2["Date"], df2["Close"], label="Close")
plt.plot(df2["Date"], df2["Predicted_Close"], label="Predicted_Close")
plt.legend()
plt.title("AAPL Stock market prediction using HMM")
plt.show()
```
