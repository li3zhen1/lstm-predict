```python
# read all png from "./visualization/LSTM_Results"
# use matplotlib to combine them into 3x2 grid

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os

# read all png from "./visualization/LSTM_Results"
path = "./visualization/LSTM_Results"
pngs = []
for f in os.listdir(path):
    if f.endswith(".png"):
        pngs.append(f)
pngs.sort()

# use matplotlib to combine them into 3x2 grid


fig = plt.figure(figsize=(50, 45))
# no padding
fig.subplots_adjust(wspace=0, hspace=0)

columns = 2
rows = 3
for i in range(1, columns*rows +1):

    img = mpimg.imread(path + "/" + pngs[i-1])


    fig.add_subplot(rows, columns, i)
    # add title
    plt.title(pngs[i-1][:-4], fontsize=30)

    # hide axis
    plt.axis('off')
    plt.imshow(img)
plt.show()
```
